# Plane Wave Operators

# 1D

```python
import numpy as np
k_min = -16
k_max = 15
k_point_grid = np.arange(k_min, k_max+1)
k_dim = len(k_point_grid)

```

```python
from grid1q.operators.plane_wave import kenetic, elec_nuc_potential, plane_wave_hamiltonian
print("k_point_grid", k_point_grid)
k_point_grid_array = [np.array([k]) for k in k_point_grid]

kenetic(k_point_grid_array)
```

```python
cell_area = 1.0
r_pos = [np.array([0.5])]

elec_nuc_potential(k_point_grid_array, cell_area, r_pos)
```

```python
H = plane_wave_hamiltonian(k_point_grid_array, cell_area, r_pos)
```

```python
from numpy.linalg import eig
eig(H)
```

# 3D
